```python
import networkx as nx
import matplotlib.pyplot as plt
import numpy as np
```

# Network node colour depends on the degree

```python
ER = nx.erdos_renyi_graph(1000, 0.01)
```

```python
degrees =  [item[1] for item in ER.degree()]
nodes = ER.nodes()
pos = nx.spring_layout(ER)
```

```python
plt.figure(figsize=(8, 6))
pos = nx.spring_layout(ER)
ec = nx.draw_networkx_edges(ER, pos, alpha=0.2)
nc = nx.draw_networkx_nodes(ER, pos, nodelist=nodes, node_color=degrees, node_size=10, cmap=plt.cm.jet)
plt.colorbar(nc)
plt.axis('off')
plt.show()
plt.savefig("")
```

# 3D

```python
Lattice = nx.grid_graph(dim=(10, 10, 10))
```

```python
xs = []
ys = []
zs = []
for item in Lattice.nodes():
    xs.append(item[0])
    ys.append(item[1])
    zs.append(item[2])
```

```python
%matplotlib notebook

# ここから可視化
fig = plt.figure(figsize=(10, 5), facecolor="w")
ax = fig.add_subplot(121, projection="3d")

# 各ノードの位置に点を打つ
ax.scatter(
    xs,
    ys,
    zs,
    s=10,
)


# エッジの表示
for e in Lattice.edges:
    node0_pos = e[0]
    node1_pos = e[1]
    xx = [node0_pos[0], node1_pos[0]]
    yy = [node0_pos[1], node1_pos[1]]
    zz = [node0_pos[2], node1_pos[2]]
    ax.plot(xx, yy, zz, c="#aaaaaa")
    
ax.axis("off")


plt.show()
```

# 2D

```python
WS = nx.watts_strogatz_graph(50, 5, p = 0.1)
```

```python
pos = {
        n: (np.cos(2*i*np.pi/50), np.sin(2*i*np.pi/50))
        for i, n in enumerate(WS.nodes)
    }
plt.figure(figsize=(10, 10))
nx.draw(WS, pos=pos, node_size = 100, width = 0.1)
```
