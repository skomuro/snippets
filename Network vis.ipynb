```python
import networkx as nx
import matplotlib.pyplot as plt
import numpy as np
```

# 3D

```python
Lattice = nx.grid_graph(dim=(10, 10, 10))
```

```python
xs = []
ys = []
zs = []
for item in Lattice.nodes():
    xs.append(item[0])
    ys.append(item[1])
    zs.append(item[2])
```

```python
%matplotlib notebook

# ã“ã“ã‹ã‚‰å¯è¦–åŒ–
fig = plt.figure(figsize=(10, 5), facecolor="w")
ax = fig.add_subplot(121, projection="3d")

# å„ãƒŽãƒ¼ãƒ‰ã®ä½ç½®ã«ç‚¹ã‚’æ‰“ã¤
ax.scatter(
    xs,
    ys,
    zs,
    s=10,
)


# ã‚¨ãƒƒã‚¸ã®è¡¨ç¤º
for e in Lattice.edges:
    node0_pos = e[0]
    node1_pos = e[1]
    xx = [node0_pos[0], node1_pos[0]]
    yy = [node0_pos[1], node1_pos[1]]
    zz = [node0_pos[2], node1_pos[2]]
    ax.plot(xx, yy, zz, c="#aaaaaa")
    
ax.axis("off")


plt.show()
```

# 2D

```python
WS = nx.watts_strogatz_graph(50, 5, p = 0.1)
```

```python
pos = {
        n: (np.cos(2*i*np.pi/50), np.sin(2*i*np.pi/50))
        for i, n in enumerate(WS.nodes)
    }
plt.figure(figsize=(10, 10))
nx.draw(WS, pos=pos, node_size = 100, width = 0.1)
```
