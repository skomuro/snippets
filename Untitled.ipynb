```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
def func(x):
    return -2 * x + 4 * x ** 3
```

```python
mu = np.linspace(1/4, 10, 100)
```

```python
x = np.sqrt(1/2 + np.sqrt(mu - 1/4))
```

```python
y = func(x)
```

```python
plt.plot(x, y)
```

```python
mu = np.linspace(1/4, 1/2, 100)
```

```python
x = np.sqrt(1/2 - np.sqrt(mu - 1/4))
```

```python
y = func(x)
```

```python
plt.plot(x, y)
```
